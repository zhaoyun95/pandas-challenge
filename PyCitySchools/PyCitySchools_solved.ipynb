```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# use the size attribute to get total schools and students
totalschools = school_data['school_name'].unique().size
totalStudents = student_data['Student ID'].unique().size

# add up budget column to get total budget
totalBudget = school_data['budget'].sum()

# straight forward calculations
avgMathScore = student_data['math_score'].sum() / totalStudents
avgReadingScore = student_data['reading_score'].sum() / totalStudents

# passing means 70 and above
passingMathCount = student_data[student_data['math_score'] >= 70]['Student ID'].size
passingMathPercent = passingMathCount / totalStudents * 100
passingReadingCount = student_data[student_data['reading_score'] >= 70]['Student ID'].size
passingReadingPercent = passingReadingCount / totalStudents * 100

# Overall Average Score
overallAverageScore = ( avgMathScore + avgReadingScore) / 2

# construction the summary dataframe
district_summary_df = pd.DataFrame({"Total Schools": [totalschools], 
                                    'Total Students': [totalStudents],
                                    'Total Budget': [totalBudget],
                                    'Average Math Score': [avgMathScore],
                                    'Average Reading Score': [avgReadingScore],
                                    '% Passing Math': [passingMathPercent],
                                    '% Passing Reading': [passingReadingPercent],
                                    'Overall Average Score': [overallAverageScore]
                                   })

# formating 
district_summary_df['Total Students'] = district_summary_df['Total Students'].map("{:,}".format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map("${:,.2f}".format)

# display results
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_summary_df = school_data.loc[:,["school_name","type",'size','budget']] 
school_summary_df.rename(columns={"school_name":"School Name", 
                                  "type":"School Type", 
                                  "size":"Total Students", 
                                  "budget":"Total School Budget"}, 
                         inplace=True)
school_summary_df["Per Student Budget"] = school_summary_df["Total School Budget"] / school_summary_df["Total Students"]

groupedBy_school =  school_data_complete.groupby('School ID')
school_summary_df["Average Math Score"] = groupedBy_school['math_score'].mean()
school_summary_df["Average Reading Score"] = groupedBy_school['reading_score'].mean()

# find students who pass math and reading
students_passing_math = school_data_complete[student_data['math_score'] >= 70]
students_passing_reading = school_data_complete[student_data['reading_score'] >= 70]

# group by column need to be consistent 'School ID'
passing_math_groupBy = students_passing_math.groupby('School ID')
passing_reading_groupBy = students_passing_reading.groupby('School ID')

school_summary_df["% Passing Math"] = passing_math_groupBy['math_score'].count() / school_summary_df["Total Students"] * 100
school_summary_df["% Passing Reading"] = passing_reading_groupBy['reading_score'].count() / school_summary_df["Total Students"] * 100
school_summary_df["% Overall Passing Rate"] = (school_summary_df["% Passing Math"] + school_summary_df["% Passing Reading"]) / 2



school_summary_df.set_index('School Name',inplace=True)

school_summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# make a copy for later use
school_summary_df_copy = school_summary_df.copy()

school_summary_df.sort_values(by='% Overall Passing Rate',inplace=True, ascending=False)

# formating
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)

school_summary_df.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
school_summary_df.sort_values(by='% Overall Passing Rate',inplace=True)

school_summary_df.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# create series for each grade
nineth_grade_df = school_data_complete[school_data_complete['grade']=='9th']
tenth_grade_df = school_data_complete[school_data_complete['grade']=='10th']
eleventh_grade_df = school_data_complete[school_data_complete['grade']=='11th']
twelveth_grade_df = school_data_complete[school_data_complete['grade']=='12th']

# greoup each series by school
nineth_grade_groupBy = nineth_grade_df.groupby('school_name')
tenth_grade_groupBy = tenth_grade_df.groupby('school_name')
eleventh_grade_groupBy = eleventh_grade_df.groupby('school_name')
twelveth_grade_groupBy = twelveth_grade_df.groupby('school_name')

# combine series into a data frame
math_scores_by_grade_df = pd.DataFrame(nineth_grade_groupBy['math_score'].mean())
math_scores_by_grade_df.rename(columns={'math_score':'9th'},inplace=True)
math_scores_by_grade_df['10th'] = tenth_grade_groupBy['math_score'].mean()
math_scores_by_grade_df['11th'] = eleventh_grade_groupBy['math_score'].mean()
math_scores_by_grade_df['12th'] = twelveth_grade_groupBy['math_score'].mean()

# display results
math_scores_by_grade_df
```

```python
# similar to above, replace math_score with reading_score
reading_scores_by_grade_df = pd.DataFrame(nineth_grade_groupBy['reading_score'].mean())
reading_scores_by_grade_df.rename(columns={'reading_score':'9th'},inplace=True)

reading_scores_by_grade_df['10th'] = tenth_grade_groupBy['reading_score'].mean()
reading_scores_by_grade_df['11th'] = eleventh_grade_groupBy['reading_score'].mean()
reading_scores_by_grade_df['12th'] = twelveth_grade_groupBy['reading_score'].mean()


reading_scores_by_grade_df
```
